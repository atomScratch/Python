```python
import sklearn
import sklearn.datasets
import seaborn as sns
import numpy as np

# Create a function to generate a subset of points on a line, with added \pm 2 noise
def generateRandomLinearSet(size, width, slope, yInt):
    def round(x): return float("{0:.2f}".format(x))
    
    outputArray = []
    for x in range(size):
        inValue = round(np.random.random_sample() * width)
        outValue = round(inValue * width + yInt + np.random.normal())
        outputArray.append([inValue,outValue])
    return outputArray

# Create the array we shall train on.
linearArray = np.array(generateRandomLinearSet(100,10,5,5))
```

```python
# Seperate all of the x-values of the array into an array with shape (n,1)
xTrain = np.reshape(linearArray[:,0],(100, 1))

# Refer to prior; substitute y-values
yTrain = np.reshape(linearArray[:,1],(100, 1))

print(xTrain)
print(yTrain)
```

```python
# Define the MyLinear object
myLinear = sklearn.linear_model.LinearRegression()

# Train the object to our training set
myLinear.fit(xTrain,yTrain)
```

```python
# Print the array the machine expects for the set we trained it on.
myLinear.predict(xTrain)
```

```python
# We "score" the program, on the training sets.
myLinear.score(xTrain,yTrain)
```

```python
# Create a new set to test the machine on
testArray = np.array(generateRandomLinearSet(100,10,5,5))

# Split into x and y arrays
xTest = np.reshape(testArray[:,0],(100, 1))
yTest = np.reshape(testArray[:,1],(100, 1))

# Score our machine on the test sets; much more realistic estimate of the accuracy
myLinear.score(xTest,yTest)
```

```python
# Predicts what yTest would be for given xTest; for user enjoyment
myLinear.predict(xTest)
```
