```python
active_file = open("20x20grid.txt", "r")

active_grid = active_file.readlines()

active_grid_strs = [string[:59].split(' ') for string in active_grid]

active_grid_ints = [[int(active_grid_strs[y_pos][x_pos]) for x_pos in range(20)] for y_pos in range(20)]
```

```python
active_grid_ints
```

```python
values_list = []

for y_pos in range(20):
    for x_pos in range(20):
        if y_pos < 17:
            values_list.append(active_grid_ints[y_pos][x_pos] * active_grid_ints[y_pos+1][x_pos] * active_grid_ints[y_pos+2][x_pos] * active_grid_ints[y_pos+3][x_pos])
            if x_pos > 2:
                values_list.append(active_grid_ints[y_pos][x_pos] * active_grid_ints[y_pos+1][x_pos-1] * active_grid_ints[y_pos+2][x_pos-2] * active_grid_ints[y_pos+3][x_pos-3])
            if x_pos < 17:
                values_list.append(active_grid_ints[y_pos][x_pos] * active_grid_ints[y_pos+1][x_pos+1] * active_grid_ints[y_pos+2][x_pos+2] * active_grid_ints[y_pos+3][x_pos+3])
        if x_pos < 17:
                values_list.append(active_grid_ints[y_pos][x_pos] * active_grid_ints[y_pos][x_pos+1] * active_grid_ints[y_pos][x_pos+2] * active_grid_ints[y_pos][x_pos+3])
```
