```python
# generating potential triples

triples_list = []

for a in range(1,1001):
    for b in range(1,1001-a):
        if a < b < 1000-(a+b):
            triples_list.append([a,b,1000-(a+b)])
```

```python
triples_list
```

```python
for val in triples_list:
    if (val[0]**2 + val[1]**2) == (val[2]**2):
        print(val)
```
